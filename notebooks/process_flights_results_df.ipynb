```python
df = pd.read_csv("../archive/2025-01-10T19-35_export.csv", index_col=0)
df
```

```python
# remove useless columns
cols_to_remove = ["delay", "airline_logo_url", "self_transfer", "hand_luggage_only"]
df = df.drop(columns=cols_to_remove)

# dtypes
df = df.loc[df["price"] != "Price unavailable"]
df["price"] = df["price"].astype(int)

# add column: price difference compared to the cheapest flight
df["price_diff"] = df["price"] - df["price"].min()

# remove duplicates
df = df.drop_duplicates()

# exclude layover time under 2 hours
# df = df.loc[df["layover_duration"] >= 120]

# reset index
df = df.reset_index(drop=True)
df
```
